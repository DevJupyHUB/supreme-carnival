```python
import matplotlib.pyplot as plt
from pyfonts import load_bunny_font
```

```python
font = load_bunny_font("Alex Brush")
```

```python
def draw_tree1(ax):       
    # Function to generate lines    
    def draw_lines(ax, x_start_base, x_end_base, y_start, y_end, num_lines, increment_x_start, increment_x_end, color="#347018", lw=2):
        for i in range(num_lines):
            # Left side
            x_start = x_start_base - i * increment_x_start
            x_end = x_end_base - i * increment_x_end
            ax.plot([x_start, x_end], [y_start, y_end], color=color, lw=lw)
            # Right side
            x_start = x_start_base + i * increment_x_start
            x_end = x_end_base + i * increment_x_end
            ax.plot([x_start, x_end], [y_start, y_end], color=color, lw=lw)

    # level 1
    draw_lines(ax, 0.5, 0.5, 0.37, 0.17, 18, 0.01, 0.02, color="#347018", lw=2)    

    #level 2
    draw_lines(ax, 0.5, 0.5, 0.56, 0.40, 13, 0.01, 0.02, color="#347018", lw=2)   

    # level 3
    draw_lines(ax, 0.5, 0.5, 0.73, 0.59, 8, 0.01, 0.02, color="#347018", lw=2)    

    # level 4
    draw_lines(ax, 0.5, 0.5, 0.91, 0.76, 6, 0, 0.02, color="#347018", lw=2)    

    # Function for plotting parallel vertical lines
    def draw_tree1_trunk(ax, num_lines, x_start, x_end, y_start, y_end, increment, color="#347018", lw=2):        
        for i in range(num_lines):
            x_start = 0.48 + i * increment
            x_end =  0.48 + i * increment
            y_start = 0.15
            y_end = 0.05
            ax.plot([x_start, x_end], [y_start, y_end], color=color, lw=lw)

    draw_tree1_trunk(ax, num_lines=5, x_start=0.48, x_end=0.48, y_start=0.15, y_end=0.05, increment=0.01, color="#347018", lw=2)

    ax.scatter(0.5, 0.95, marker="*", s=500, color="#F2C808")
    
    ax.set_aspect("equal") 
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)     
```

```python
def draw_tree2(ax):       
    # Function to generate lines 
    def draw_lines(ax, num_lines=5, y_start=0.91, y_spacing=0.02, length_start=0.1, length_increment=0.05, color="#347018", lw=3):
        for i in range(num_lines):            
            y_position = y_start - i * y_spacing            
            line_length = length_start + i * length_increment            
            x_start = 0.5 - line_length / 2              
            x_end = 0.5 + line_length / 2              
            ax.plot([x_start, x_end], [y_position, y_position], color=color, lw=lw)

    # level 1
    draw_lines(ax, num_lines=8, y_start=0.36, y_spacing=0.02, length_start=0.4, length_increment=0.05)    

    #level 2
    draw_lines(ax, num_lines=8, y_start=0.54, y_spacing=0.02, length_start=0.3, length_increment=0.05)   

    # level 3
    draw_lines(ax, num_lines=8, y_start=0.72, y_spacing=0.02, length_start=0.2, length_increment=0.05)    

    # level 4
    draw_lines(ax, num_lines=8, y_start=0.90, y_spacing=0.02, length_start=0.1, length_increment=0.05)       
    
    # Function for plotting parallel horizontal lines
    def draw_tree2_trunk(ax, num_lines=5, y_start=0.9, y_spacing=0.05, line_length=0.5, color="#347018", lw=3):
        for i in range(num_lines):            
            y_position = y_start - i * y_spacing            
            x_start = 0.5 - line_length / 2  
            x_end = 0.5 + line_length / 2                
            ax.plot([x_start, x_end], [y_position, y_position], color=color, lw=lw)
    
    draw_tree2_trunk(ax, num_lines=5, y_start=0.18, y_spacing=0.02, line_length=0.1, color="#347018", lw=3)

    ax.scatter(0.5, 0.95, marker="*", s=500, color="#F2C808")

    ax.set_aspect("equal")    
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)      
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
plt.subplots_adjust(wspace=0.0)

draw_tree1(ax[0])  
draw_tree2(ax[1]) 

fig.suptitle("Merry Christmas!", fontsize=50, font=font, y=1.05, color="#347018")

ax[0].axis("off")
ax[1].axis("off")

fig.set_facecolor("#fefefc")

fig.text(0.1, 0.08, "Created with Matplotlib | Design: Bernadett Piros", color="#347018", font=font, size=10, alpha=0.8)

plt.show()

#fig.savefig("Merry_Xmas.png", bbox_inches="tight", pad_inches=0.2)
```
