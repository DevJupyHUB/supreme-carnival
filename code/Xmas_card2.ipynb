```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Circle, Rectangle, Polygon
from pyfonts import load_font

# load font 
font = load_font(
   font_url='https://github.com/google/fonts/blob/main/ofl/overtherainbow/OvertheRainbow.ttf?raw=true'
)

# Set up the figure
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

# Background and snow on the ground
fig.patch.set_facecolor('#191970')
ax.add_patch(Rectangle((0, 0), 1, 0.2, color='#f6f6f6'))

# Snowflakes
n_snowflakes = 200
snowflakes_x = np.random.uniform(0, 1, n_snowflakes)
snowflakes_y = np.random.uniform(0.2, 1, n_snowflakes)
ax.scatter(snowflakes_x, snowflakes_y, color='white', s=10, marker='*')

# Draw tree
# Trunk
ax.add_patch(Rectangle((0.25, 0.15), 0.08, 0.15, color='#654321')) 
# Bough
tree_levels = [2, 1.5, 1]
start = 0.65
step = 0.5
d = 0.3
for level, scale in zip(np.arange(start, start + step * len(tree_levels), step), tree_levels):
    ax.plot(0.3, 0.15 + scale * d, marker='^', markersize=(4 - scale) * 50, color='green')
# Star   
ax.plot(0.3, 0.15 + 2.4 * d, marker='*', markersize=30, color='yellow')  

# Baubles
baubles = [
    (0.3, 0.5, 0.15), (-0.4, 0.45, 0.1), (0.2, 1.0, 0.15),
    (0.1, 1.4, 0.1), (0.0, 2.0, 0.15), (-0.15, 1.7, 0.15),
    (0.15, 0.65, 0.1), (-0.3, 1, 0.1), (-0.05, 0.85, 0.15),
    (-0.2, 0.55, 0.1), (-0.2, 1.3, 0.15), (0.05, 1.1, 0.1),
    (0.15, 1.75, 0.15)
]
for dx, dy, scale in baubles:
    ax.plot(0.3 + dx * d, 0.18 + dy * d, marker='o', markersize=scale * 100, color='red')

# Draw snowman
snowman_parts = [Circle((0.75, 0.25), 0.1), Circle((0.75, 0.4), 0.075)]
for part in snowman_parts:
    part.set_color('white')
    ax.add_patch(part)

# Buttons and eyes
buttons_and_eyes = [(0.75, 0.22), (0.75, 0.26), (0.75, 0.3), (0.72, 0.41), (0.78, 0.41)]
ax.scatter(*zip(*buttons_and_eyes), color='black', s=np.random.uniform(20, 45, 5))

# Arms, nose, and hat
ax.plot([0.67, 0.58], [0.3, 0.33], color='#654321', lw=3)
ax.plot([0.83, 0.9], [0.3, 0.34], color='#654321', lw=3)
nose_pts = [[0.75, 0.4], [0.75, 0.37], [0.79, 0.385]]
hat_pts = [[0.68, 0.45], [0.75, 0.57], [0.82, 0.45]]
for shape, color in [(nose_pts, 'orange'), (hat_pts, 'red')]:
    ax.add_patch(Polygon(shape, closed=True, color=color))
ax.plot(0.75, 0.58, marker='o', markersize=18, color='white')  # Tassel

# Message 
ax.text(0.65, 0.08, 'Merry Christmas!', color='red', font=font, fontsize=20, fontweight='bold', ha='center')

ax.axis('off')
plt.show()
```
